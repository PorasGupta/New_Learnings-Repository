# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib.pyplot as plt

Q= """
SELECT * from master
"""
df = pd.read_sql_query(Q,conn)

plt.hist(df['ConvertedComp'])
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
q = """ SELECT * FROM master """
df = pd.read_sql_query(q,conn)

df['Age'].plot(kind = 'box')
plt.title('Box plot for Age of Respondents')
plt.ylabel('Age Intervals')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

df.plot(x = 'Age', y = 'WorkWeekHrs', kind = 'scatter', figsize = (14,6), color = 'green')
plt.title('Relation Between Age and Work Week Hours of Respondents')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
normAge = (df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())
df.plot(kind = 'scatter', x= 'WorkWeekHrs', y='CodeRevHrs', alpha = 0.5, s = normAge*1000 + 10, figsize = (14,6), color = 'darkred')

plt.title('Relation Between WorkWeekHrs and CodeRevHrs of Respondents')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.legend(['Age'], loc='upper right', fontsize='x-large')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

q = """ Select Count(Respondent) as Total_Respondent, DatabaseDesireNextYear from DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by Total_Respondent desc limit 5
"""
df_pie = pd.read_sql_query(q, conn)
df_pie.set_index('DatabaseDesireNextYear', inplace = True)

df_pie.plot(kind = 'pie', y = 'Total_Respondent', autopct = '%1.1f%%', startangle = 90, figsize = (10,10))
plt.legend()
plt.title('Top 5 Databases that Respondents Wish to Learn Next Year')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.ylabel('')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
q = """
Select WorkWeekHrs, CodeRevHrs, Age
from master
where Age >= 30 and Age <=35
"""
df_stack = pd.read_sql_query(q, conn)
df_stack.groupby(['CodeRevHrs', 'Age'])['WorkWeekHrs'].median().sort_values(ascending=False).unstack().plot(kind = 'bar', figsize = (25,10), stacked = True)

plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs For The Age Group 30 to 35',
          loc='center',pad=30, fontsize=30, color='blue')
plt.xlabel('CodeRevHrs', fontsize = 20)
plt.ylabel('WorkWeekHrs', fontsize = 20)
plt.legend(bbox_to_anchor=(1, 1), shadow=True, ncol=1, title='Age')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
q = """
Select Age, ConvertedComp
from master
where Age >=45 and Age <=60
"""
df1 = pd.read_sql_query(q, conn)
df_line = df1.groupby('Age')['ConvertedComp'].median().reset_index()
df_line.head(20)

df_line.plot(kind = 'line', x = 'Age', y = 'ConvertedComp', color = 'y', figsize = (15,6))

plt.title('Average Compensation Based on Age Groups')
plt.ylabel('Converted Compensation (Annual USD)')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as MainBranchCount
from master
group by MainBranch
"""

df_hbar=pd.read_sql_query(QUERY,conn)

df_hbar.plot(kind='barh', figsize=(10,6), color='lightskyblue')

plt.xlabel('MainBranchCounts')
plt.ylabel('MainBranchType')
plt.text(1000, 1, 'I am a Developer')
plt.text(1000, 0, 'I am a Developer, but only write code sometimes')

plt.show()
```
